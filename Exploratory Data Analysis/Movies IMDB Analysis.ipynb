```python
filtered_df = df[['Country', 'Runtime','Directors']]
filtered_df.head()
```

##### Find the number of movies in netflix with IMDb score greater than 7.5

```python
df[(df['Platform'] == 'Netflix') & (df['IMDb'] > 7.5)].shape[0]
```

##### Find the number of movies in comedy genre or with runtime lesser than 120

```python
df[(df['Genres'] == 'Comedy') | (df['Runtime'] < 120)].shape[0]
```

##### Find the average imdb score for each genres

```python
avg_imdb = df.groupby('Genres')[['IMDb']].mean().rename(columns ={'IMDb':'Average IMDb'}).reset_index()
avg_imdb
```

#### find the avg rotten tomatoes score for each platform


```python
avg_rotten = df.groupby('Platform')[['Rotten Tomatoes']].mean().rename(columns ={'Rotten Tomatoes':'Average Rotten Tomatoes'}).reset_index()
avg_rotten
```

#### Find the avg runtime of each genre on each platform

```python
avg_runtime = df.groupby(['Genres','Platform'])['Runtime'].mean().rename('Average Runtime').reset_index()
avg_runtime
```

##### Find the avg runtime and avg imdb score for each platform

```python
avg_runtime_platform = df.groupby('Platform')[['Runtime','IMDb']].mean().rename(
                       columns = {'Runtime':'Average Runtime','IMDb':'Average IMDb'}).reset_index().round(2)
avg_runtime_platform
```

#### Find the total runtime and avg imdb score for each platform.

```python
runtime_imdb_platform = df.groupby('Platform').agg({"Runtime":"sum","IMDb":"mean"}).rename(
                       columns = {'Runtime':'Total Runtime','IMDb':'Average IMDb'}).reset_index().round(2)
runtime_imdb_platform
```

#### Top 3 languages  based on IMdb Score

```python
new_df = df[["Language","IMDb"]]
top_3_languages = new_df.nlargest(3, "IMDb")
top_3_languages
```
